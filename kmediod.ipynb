```python
import numpy as np

X = np.array([
    [2, 6],
    [3, 4],
    [3, 8],
    [4, 7],
    [6, 2],
    [6, 4],
    [7, 3],
    [7, 4]
])

k = 2

# Step 1: initialize medoids (pick first k points)
medoids = X[:k]

def manhattan_distance(a, b):
    return np.sum(np.abs(a - b))

for iteration in range(10):
    # Step 2: assign points to nearest medoid
    clusters = {i: [] for i in range(k)}
    for point in X:
        distances = [manhattan_distance(point, m) for m in medoids]
        cluster_id = np.argmin(distances)
        clusters[cluster_id].append(point)

    # Step 3: update medoids
    new_medoids = []
    for cluster in clusters.values():
        cluster = np.array(cluster)
        costs = []
        for candidate in cluster:
            cost = sum(manhattan_distance(candidate, p) for p in cluster)
            costs.append(cost)
        new_medoids.append(cluster[np.argmin(costs)])

    new_medoids = np.array(new_medoids)

    if np.array_equal(medoids, new_medoids):
        break

    medoids = new_medoids

print("Final Medoids:")
print(medoids)

print("\nClusters:")
for i, cluster in clusters.items():
    print(f"Cluster {i+1}: {cluster}")
```
